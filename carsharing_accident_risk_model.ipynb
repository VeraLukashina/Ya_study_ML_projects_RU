3. Понять, помогут ли результаты моделирования и анализ важности факторов ответить на вопросы:
    * Возможно ли создать адекватную системы оценки водительского риска при выдаче авто?

### Анализ распределения количества ДТП по месяцам.

Вычислим количество ДТП по годам.

```python
query = '''
SELECT 
    DATE_TRUNC('year', collision_date)::date AS year_date,
    COUNT(case_id) AS accidents      
FROM
    collisions
GROUP BY 
    year_date;
'''
year_collisions = pd.read_sql_query(query, con=con)
year_collisions.head()
```

Последняя запись в таблице датируется 2013 годом, но, судя по колличеству записей, 2012 год также представлен не полностью. Выведем количество ДТП за 2012 год по месяцам.

```python
query = '''
SELECT 
    DATE_TRUNC('month', collision_date)::date AS month_date,
    COUNT(case_id) AS accidents      
FROM
    collisions
WHERE
    collision_date BETWEEN '2012-01-01' AND '2012-12-31'
GROUP BY 
    month_date;
'''
month_year = pd.read_sql_query(query, con=con)
month_year
```

Начиная с мая 2012 количество записей в ДБ существенно снижается. Поэтому для статистического анализа будем использовать только данные за 3 года: с 01.01.2009 по 31.12.2011

Рассчитаем среднее количество аварий в месяц за указанный период.

```python
query = '''
WITH acc_per_month AS (
    SELECT 
        DATE_TRUNC('month', collision_date)::date AS month_date,
        COUNT(case_id) AS accidents      
    FROM
        collisions
    WHERE 
        collision_date BETWEEN '2009-01-01' AND '2011-12-31'
    GROUP BY 
        month_date
)
  
SELECT 
    EXTRACT(MONTH FROM month_date)::int AS month_of_accident,
    AVG(accidents) AS accidents_mean
FROM 
    acc_per_month
GROUP BY    
    month_of_accident;
'''
data_collisions = pd.read_sql_query(query, con=con)
data_collisions
```

```python
plt.figure()
plt.bar(data_collisions.month_of_accident, data_collisions.accidents_mean/1000)
plt.xlabel('Номер месяца')
plt.ylabel('Количество ДТП, тыс.')
plt.title("Распределение количества ДТП по месяцам")
plt.grid(axis='y')
plt.show()
```

Наибольшее количество аварий наблюдается в октябре, в то время как в феврале их число минимально. Это можно объяснить изменением погодных условий, которые с октября становятся особенно неблагоприятными, включая плохое состояние дорожного покрытия и недостаточное освещение. Кроме того, водители могут еще не успеть приспособиться к новым условиям после летних месяцев, что также способствует увеличению числа ДТП.

Построим график, отражающий помесячное количество аварий за исследуемый период.

```python
query = '''
SELECT 
    DATE_TRUNC('month', collision_date)::date AS month_date,
    COUNT(case_id) AS accidents      
FROM
    collisions
WHERE 
    collision_date BETWEEN '2009-01-01' AND '2011-12-31'
GROUP BY 
    month_date;

'''
df_collisions = pd.read_sql_query(query, con=con, parse_dates=['month_date'])
df_collisions.head()
```

```python
plt.plot(df_collisions['month_date'], df_collisions['accidents']/1000)
# аварии в октябре
october = df_collisions.loc[df_collisions['month_date'].dt.month == 10]
plt.scatter(x=october['month_date'], y=october['accidents']/1000, color='r', label='Октябрь')
# аварии в феврале
february = df_collisions.loc[df_collisions['month_date'].dt.month == 2]
plt.scatter(x=february['month_date'], y=february['accidents']/1000, color='g', label='Февраль')
# Заголовок, метка x и метка y
plt.title('Динамика количества ДТП')
plt.xlabel('Дата ДТП')
plt.xticks(rotation=45)
plt.ylabel('Количество ДТП, тыс.')
plt.legend()
plt.show()
```

График динамики количества ДТП за 3 года так же демонстрирует, что, несмотря на общий тренд к снижению количества аварий, наиболее опасный месяц - это октябрь. Февраль же, несмотря на самое низкое среднее значение, в динамике не является наиболее безопасным месяцем.

Проанализируем влияние таких факторов как погодные условия, состояние дорожного покрытия и уровень освещения на среднее значение количества ДТП по месяцам.

Объявим функции для выгрузки и визуализации соответствующих данных.

```python
def monthly_details(col_name):
    '''
    Рассчитывает среднее количество ДТП в месяц в разрезе указанного фактора за период 2009-2011 включительно.
    Принимает на вход назавание столбца из таблицы collisions.
    Возвращает датафрейм   
    '''
    query = f'''
        WITH acc_per_month AS (
        SELECT 
            DATE_TRUNC('month', collision_date)::date AS month_date,
            {col_name},
            COUNT(case_id) AS accidents      
        FROM
            collisions
        WHERE collision_date BETWEEN '2009-01-01' AND '2011-12-31'
        GROUP BY 
            {col_name},
            month_date
    )  
    SELECT 
        EXTRACT(MONTH FROM month_date)::int AS month_of_accident,
        {col_name},
        AVG(accidents) AS accidents_mean
    FROM 
        acc_per_month
    GROUP BY    
        month_of_accident,
        {col_name}
    ORDER BY
        month_of_accident;
    '''
    return pd.read_sql_query(query, con=con)
```

```python
def draw_monthly_details(data, col_name, title):
    plt.figure()
    sns.barplot(data=data, x='month_of_accident', y='accidents_mean', hue=col_name)
    plt.xlabel('Номер месяца')
    plt.ylabel('Количество ДТП')
    plt.title(f"Распределение количества ДТП по месяцам в зависимости от {title}")
    plt.grid(axis='y')
    plt.show()
```

В летние месяцы количество аварий под влиянием неблагоприятных факторов значительно ниже по сравнению с другими сезонами. Это вполне логично, учитывая, что летом погодные условия чаще благоприятны, а продолжительность светового дня больше.

Повышенное количество ДТП в октябре не объясняется исключительно погодными условиями, возможно, в октябре увеличивается общее количество поездок.

### Анализ влияния факторов, понижающих внимание водителя, на количество ДТП.

Алгоритм решения задачи:
1. Определение факторов, понижающих внимание водителя:
    * алкогольное или наркотическое опьянение (parties-party_drug_physical, parties-party_sobriety),
    * использование телефона во время вождения (parties-cellphone_in_use),
    * усталость/сонливость (parties-party_drug_physical и collisions-pcf_violation_category) 
    
2. Написание SQL-запроса для каждого фактора.

3. Построение графика для каждого фактора.

```python
def accident_pie(X, labels, title):
    plt.figure()
    plt.pie(X, labels=labels, autopct='%.1f%%', startangle=140)
    plt.title(title)
    plt.show()
```

```python
query = '''
WITH all_cars AS (
    SELECT 
        COUNT(DISTINCT(case_id)) AS count_all
    FROM 
        parties 
    WHERE 
        party_type='car'
),
    alcohol_or_drug AS (
    SELECT 
        COUNT(DISTINCT(case_id)) count_alcohol_or_drug
    FROM 
        parties 
    WHERE 
        party_type='car' AND (
        party_sobriety IN ('had been drinking, under influence', 
                           'had been drinking, not under influence', 
                           'had been drinking, impairment unknown') 
        OR party_drug_physical = 'under drug influence')
)
SELECT
    all_cars.count_all - ad.count_alcohol_or_drug AS no_alcohol_or_drug,
    ad.count_alcohol_or_drug AS alcohol_or_drug
FROM 
    all_cars 
    FULL JOIN alcohol_or_drug AS ad ON true;
'''
alcohol_or_drug = pd.read_sql_query(query, con=con)
alcohol_or_drug = alcohol_or_drug.T
alcohol_or_drug
```

```python
accident_pie(alcohol_or_drug[0], 
          ['Иное состояние', 'Состояние алкогольного или наркотического опьянения'],
          'Доля ДТП, в которых водитель находился в состоянии алкогольного или наркотического опьянения')
```

```python
query = '''
WITH phone AS (
    SELECT
        DISTINCT(case_id) unique_case_id,
        SUM(cellphone_in_use) num_phones
    FROM
        parties
    WHERE
        party_type='car'
    GROUP BY
        case_id
),
    acc_phone_in_use AS (
    SELECT
        unique_case_id,
        CASE
            WHEN num_phones >= 1 THEN 1
            ELSE 0        
        END AS cellphone_in_use
    FROM
        phone
)  
SELECT
    COUNT(unique_case_id) - SUM(cellphone_in_use) AS no_cellphone,
    SUM(cellphone_in_use)  AS cellphone_in_use
FROM
    acc_phone_in_use;
'''
cellphone_in_use = pd.read_sql_query(query, con=con)
cellphone_in_use = cellphone_in_use.T
cellphone_in_use
```

```python
accident_pie(cellphone_in_use[0], 
             ['Телефон не использовался', 'Телефон использовался'],
             'Доля ДТП, в которых водитель пользовался телефоном')
```

```python
query = '''
WITH fell_asleep AS (
    SELECT
        COUNT(DISTINCT(p.case_id)) AS count_fell_asleep
    FROM
        parties AS p
        JOIN collisions AS c ON p.case_id=c.case_id
    WHERE
        p.party_type='car' 
        AND
        p.party_drug_physical='sleepy/fatigued'
        AND
        c.pcf_violation_category='fell asleep'
),    
    sleepy_fatigued AS (
    SELECT
        COUNT(DISTINCT(case_id)) AS count_sleepy_fatigued
    FROM
        parties
    WHERE
        party_type='car' 
        AND
        party_drug_physical='sleepy/fatigued'
),    
    all_cars AS (
    SELECT
        COUNT(DISTINCT(case_id)) AS count_all
    FROM
        parties
    WHERE
        party_type='car'
)
SELECT
    all_cars.count_all - sleepy_fatigued.count_sleepy_fatigued AS no_sleepy_fatigued,
    sleepy_fatigued.count_sleepy_fatigued - fell_asleep.count_fell_asleep AS sleepy_fatigued,
    fell_asleep.count_fell_asleep AS fell_asleep
FROM
    all_cars
    JOIN sleepy_fatigued ON true
    JOIN fell_asleep ON true;
    
'''
sleep = pd.read_sql_query(query, con=con)
sleep = sleep.T
sleep
```

```python
accident_pie(sleep[0],
            ["Иное состояние", "Состояние сонливости или усталости",""],
            "Доля ДТП, в которых водитель находился в состоянии сонливости или усталости")
```

Из факторов, понижающих внимание водителя, наибольшая доля ДТП приходится на вождение в состоянии алкогольного или наркотического опьянения (около 11%). Использование телефона было зафиксировано менее чем в 2,5%, а доля аварий с участием уставших или сонных водителей составляет менее 1% от всех ДТП за исследуемый период.

### Исследовать связь между типом коробки передач транспортного средства и количеством спровоцированных ДТП.

```python
query = '''
SELECT    
    v.vehicle_transmission,
    p.at_fault,
    COUNT(DISTINCT(v.case_id, v.party_number)) AS num_of_accidents
FROM 
    vehicles v
    JOIN parties p ON v.case_id=p.case_id AND v.party_number = p.party_number
WHERE
    v.vehicle_transmission IN ('auto', 'manual')
GROUP BY 
    v.vehicle_transmission,
    p.at_fault;
'''
transmission = pd.read_sql_query(query, con=con)
transmission
```

```python
# Данные
fault = ["Не виновен", "Виновен"]
transmission_auto = transmission.loc[transmission['vehicle_transmission'] == 'auto', 'num_of_accidents']
transmission_manual = transmission.loc[transmission['vehicle_transmission'] == 'manual', 'num_of_accidents']

# Ширина каждого столбца
bar_width = 0.3

# Позиции для каждой группы столбцов
r1 = range(len(transmission_auto))
r2 = [x + bar_width for x in r1]

# Построение столбцов для каждой группы
plt.bar(r1, transmission_auto, width=bar_width, edgecolor='grey', label='Автоматическая')
plt.bar(r2, transmission_manual, width=bar_width, edgecolor='grey', label='Ручная')

plt.xticks([r + bar_width/2 for r in range(len(transmission_auto))], fault)
plt.ylabel('Количество ДТП', fontweight='bold')
plt.legend()
plt.title('Количество ДТП в разрезе типов КПП и виновности водителя')
plt.show()
```

Водители ТС с автоматической КПП реже являются виновниками аварии, но при этом чаще попадают в ДТП, спровоцированные другими участниками.

**Выводы:**

* Текущие погодные условия и степень освещенности играют более значительную роль в вероятности возникновения ДТП, чем дата поездки.

* Состояние алкогольного или наркотического опьянения с наибольшей вероятностью приведет к ДТП, нежели разговор по телефону или усталость.

* Виновником аварии чаще становится водитель ТС с ручной КПП.

## Разработка модели машинного обучения.


Для оценки модели в данной бизнес-задаче будет использоваться метрика Recall. Она измеряет долю истино-положительных результатов относительно всех реально положительных случаев.

Обоснование выбора:

Согласно описанию, в случае если система определила высокий уровень риска ДТП на выбранном водителем маршруте, он должен увидеть предупреждение и рекомендации по маршруту. Ошибка при присвоении маршруту метки аварийноопасного принесет меньше потерь, чем пропуск действительно опасного маршрута. Это означает, система должна отдавать приоритет количественной, а не качественной мере. Этому требованию отвечает метрика recall.


### LogisticRegression

Для корректной работы модели Логистической регрессии числовые данные должны быть отмасштабированы, категориальные - закодированы.

```python
# Обработка категориальных признаков
ohe_pipe = make_pipeline(OneHotEncoder(drop='if_binary', handle_unknown='ignore', sparse_output=False))

cbe_pipe = make_pipeline(CatBoostEncoder(random_state=SEED))

# Обработка числовых признаков
numeric_pipe = make_pipeline(StandardScaler())

# Трансформер для всех признаков
full = ColumnTransformer(
    transformers=[
        ('small_cat', ohe_pipe, x_small_cat),
        ('big_cat', cbe_pipe, x_big_cat),
        ('standardization', numeric_pipe, x_num)
    ],
    n_jobs=-1
)
```

```python
def optimaze_logreg(trial):
    C = trial.suggest_float("C", 0.001, 10)
    solver = trial.suggest_categorical("solver", ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'])
    
    logreg = Pipeline(steps=[('preprocess', full),
                             ('model_lr', LogisticRegression(C=C, random_state=SEED, solver=solver,
                                                              max_iter=350, n_jobs=-1))])
    
    logreg.fit(x_train, y_train)
    preds = logreg.predict(x_valid)
    score = recall_score(y_valid, preds)
    return score
```

```python
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size_1, hidden_size_2, hidden_size_3, output_size):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size_1) 
        self.tn1 = nn.Tanh()
        self.fc2 = nn.Linear(hidden_size_1, hidden_size_2)
        self.relu = nn.ReLU()
        self.fc3 = nn.Linear(hidden_size_2, hidden_size_3)
        self.relu2 = nn.ReLU()
        self.fc4 = nn.Linear(hidden_size_3, output_size)
        self.sg = nn.Sigmoid()
                       
   
    def forward(self, x):
        out = self.fc1(x)
        out = self.tn1(out)
        out = self.fc2(out)
        out = self.relu(out)
        out = self.fc3(out)
        out = self.relu2(out)
        out = self.fc4(out)
        out = self.sg(out)
        return out
```

Объявим функцию для обучения и оценки нейронной сети.

```python
def neuralnet_train_eval(model, train_loader, valid_loader, num_epochs):
    '''
    Обучает модель на тренировочной выборке, оценивает на валидационной.
    Выводит значение Recall каждые 250 эпох.
    Принимает на вход:
    * модель,
    * DataLoader с тренировочной выборкой,
    * DataLoader с валидационной выборкой,
    * общее количество эпох
    Возвращает обученную модель, получившую лучшую оценку на валидационной выборке.
    '''
    best_score = 0
        
    # Обучение модели
    for epoch in range(num_epochs):
        model.train()
        all_train_outputs = []
        all_train_targets = []
        for inputs, targets in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets.unsqueeze(1))

            loss.backward()
            optimizer.step()
            
            all_train_outputs.append(outputs)
            all_train_targets.append(targets)
            
        if (epoch+1) % 250 == 0 or (epoch+1) == num_epochs :
            all_train_outputs = torch.cat(all_train_outputs)
            all_train_targets = torch.cat(all_train_targets)
            all_train_targets = all_train_targets.long()
            net_score_train = binary_recall(all_train_outputs.squeeze(), all_train_targets.squeeze())
            print(f'Эпоха {(epoch+1)}/{num_epochs}')
            print(f'Значение Recall на трейне {net_score_train:.3f}')               

    # Оценка модели на валидационном наборе данных
            model.eval()
            all_valid_outputs = []
            all_valid_targets = []

            with torch.no_grad():
                for inputs, targets in valid_loader:
                    outputs = model(inputs)
                    all_valid_outputs.extend(outputs)  
                    all_valid_targets.extend(targets)
            
            all_valid_outputs = torch.cat(all_valid_outputs)
            all_valid_targets = torch.stack(all_valid_targets)
            all_valid_targets = all_valid_targets.long()
            net_score_valid = binary_recall(all_valid_outputs.squeeze(), all_valid_targets.squeeze())
            print(f'Значение Recall на валидации {net_score_valid:.3f}')
            
            if net_score_valid > best_score:
                best_score = net_score_valid
                best_model = model
                
    return best_model
```

```python
class NeuralNetReg(nn.Module):
    def __init__(self, input_size, hidden_size_1, hidden_size_2, hidden_size_3, output_size):
        super(NeuralNetReg, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size_1)
        self.tn1 = nn.Tanh()
        self.dp1 = nn.Dropout(p=0.6)
        self.fc2 = nn.Linear(hidden_size_1, hidden_size_2)
        self.relu = nn.ReLU()
        self.fc3 = nn.Linear(hidden_size_2, hidden_size_3)
        self.relu2 = nn.ReLU()
        self.fc4 = nn.Linear(hidden_size_3, output_size)
        self.sg = nn.Sigmoid()
                       
   
    def forward(self, x):
        out = self.fc1(x)
        out = self.tn1(out)
        out = self.dp1(out)
        out = self.fc2(out)
        out = self.relu(out)
        out = self.fc3(out)
        out = self.relu2(out)
        out = self.fc4(out)
        out = self.sg(out)
        return out
```

Рассчитаем время выполнения предсказаний каждой из нейросетей.

```python
def net_predict(model, data_loader):
    '''
    Рассчитывает качество и время работы модели в режиме предказания.
    Принимает на вход:
    * модель,
    * DataLoader с выборкой,
    Возвращает recall и продолжительность работы.
    '''
    start_time_pred = timeit.default_timer()
    model.eval()
    all_valid_outputs = []
    all_valid_targets = []

    with torch.no_grad():
        for inputs, targets in valid_loader:
            outputs = model(inputs)
            all_valid_outputs.extend(outputs)  
            all_valid_targets.extend(targets)
            
    all_valid_outputs = torch.cat(all_valid_outputs)
    all_valid_targets = torch.stack(all_valid_targets)
    all_valid_targets = all_valid_targets.long()
    net_score_valid = binary_recall(all_valid_outputs.squeeze(), all_valid_targets.squeeze())

    end_time_pred = timeit.default_timer()
    time_pred = end_time_pred - start_time_pred
    
    return net_score_valid, time_pred
```
